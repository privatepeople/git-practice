# 라이브러리

```python
import numpy as np
import pandas as pd
import scipy as sp
from scipy import stats
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
%matplotlib inline
```

```python
# 감자칩의 데이터
potato_chip = np.array([89.03, 95.08, 88.26, 90.07, 90.6, 87, 87.67, 88.8, 90.46, 81.33])
```

```python
one_sample = stats.ttest_1samp(potato_chip, popmean=90)
```

```python
t_statistic, pvalue = one_sample
```

```python
pvalue
```

```python
ax = sns.kdeplot(x=potato_chip, label="potato_chip")
ax.set(xlabel="The weight for the snack")
plt.legend()
plt.show()
```
